```python
import tensorflow as tf
```

We can load **The Fashion MNIST data** directly from the tf.keras datasets API:

```python
mnist = tf.keras.datasets.fashion_mnist
```

The Fashion MNIST is a dataset comprised of 60,000 small square 28Ã—28 pixel grayscale images of items of 10 types of clothing. The mapping of all 0-9 integers to class labels is listed below:


0: T-shirt/top

1: Trouser

2: Pullover

3: Dress

4: Coat

5: Sandal

6: Shirt

7: Sneaker

8: Bag

9: Ankle boot

We obtain the training and testing values for the graphics that contain the clothing items and their labels by calling **load_data** :

```python
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
```

Now Let's take an example to print a training image, and a training label...

```python
import numpy as np
np.set_printoptions(linewidth=200)
import matplotlib.pyplot as plt
plt.imshow(training_images[0])
print(training_labels[0])
print(training_images[0])
```

```python
np.set_printoptions(linewidth=200)
plt.imshow(training_images[13])
print(training_labels[13])
print(training_images[13])
```

```python
np.set_printoptions(linewidth=200)
plt.imshow(training_images[26])
print(training_labels[26])
print(training_images[26])
```

We notice that all of the values in the number are between 0 and 255.
A process called **normalizing** in Python it's easy to normalize a list to treat all values as between 0 and 1, without looping (it's easier in training a neural network). :

```python
training_images  = training_images / 255.0
test_images = test_images / 255.0
```

Let's now design the model!

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(), 
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu), 
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
from tensorflow.keras import optimizers
optimizers.Adam()
```

```python
model.compile(optimizer = optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(training_images, training_labels, epochs=5)
```

```python
model.evaluate(test_images, test_labels)
```
